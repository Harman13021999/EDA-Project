```python
# Data Cleaning

# England had missing country code so to fill up the country code with ENG string.
df1.loc[df1['Country'] == 'England','Country_Code'] = 'ENG'
df1.loc[df1['Country'] == 'England']

```

```python
# Removing the Country whose value is a Continent name or Region name.
df1.dropna(subset = ['Country_Code'], inplace = True)
```

```python
# The following query removes particular rows from the Country column. 
df1.drop(df1.index[df1['Country'] == 'World'], inplace=True)
df1.drop(df1.index[df1['Country'] == 'Western Sahara'], inplace=True)
```

```python
# Removing the Nan Values messed up the indexes so resetting the indexes.
df1.reset_index(drop = True, inplace = True)
```

```python
# Replacing Nan values with 0 for whole dataframe. 
df1.fillna(value=0, inplace=True)
```

```python
# Changing the data type of column with float64 to int64.
for column in df1.columns:
    if df1[column].dtype == 'float64':
        df1[column] = df1[column].astype('int64')
```
